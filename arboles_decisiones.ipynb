```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
df=sns.load_dataset("iris")
df
```

```python
df['species'].unique()
```

Cuando quiero identificar a que categoria es usamos arboles de decisiones

```python
# las variables independientes vana a ser todos menos las de las especies
#eliminamos la columna especies
X=df.drop('species',axis=1)
X
```

```python
# la variable dependiente es el tipo de especie que yo quiero saber de la flor
le=LabelEncoder()
especies=le.fit_transform(df['species'])
especies
```

```python
y=especies
X_entrena,X_prueba,y_entrena,y_prueba=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
arbol=DecisionTreeClassifier()
```

```python
arbol.fit(X_entrena,y_entrena)
```

```python
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol);
```

```python
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol,class_names=['setosa', 'versicolor', 'virginica'],filled=True);
# x[2] es de las columnas que se utilizo , para eso se agrega un valor mas a la funcion
```

```python
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol,class_names=['setosa', 'versicolor', 'virginica'],feature_names=df.columns.to_list(),filled=True);
```
