```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
ruta=r"/content/datos_seguror.csv"
df=pd.read_csv(ruta)
df
```

```python
plt.scatter(df.edad,df.compra)
```

```python
# separa dos grupos uno para prueba y otro para entrenamiento
X_entrena,X_prueba,y_entrena,y_prueba=train_test_split(df[["edad"]],df['compra'],train_size=0.9)
```

```python
#en la celda anterior ya dividio los datos en cada una de las variables
X_entrena
```

```python
X_prueba
```

En x_entrena tiene mayor datos , ya que yo puse en train_size que entrene con el 90% de los datos....

```python
modelo=LogisticRegression()
modelo.fit(X_entrena,y_entrena)
```

```python
modelo.score(X_entrena,y_entrena)
```

se obtuvo un procentaje de 0.9166 , lo cual es muy bueno

```python
datos_nuevos=pd.DataFrame({"edad":[25,35,45,55]})
datos_nuevos
```

```python
probabilidades=modelo.predict_proba(datos_nuevos)
probabilidades
```

```python
# solo la segunda columna es la que me importa , ya que es la probabilidad que me va comprar el seguro
probabilidades=probabilidades[:,1]
probabilidades
```
