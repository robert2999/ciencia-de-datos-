```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model 
# Crear DataFrame
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)
X=df[['Edad']]
y=df['Salario']
modelo=linear_model.LinearRegression()
modelo.fit(X,y)
plt.plot(df["Edad"],modelo.predict(X))
plt.title("Relación entre Edad y Salario")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.grid(True);
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import linear_model
# Crear DataFrame
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)
X=df[["Edad"]]
y=df['Salario']
modelo=linear_model.LinearRegression()
modelo.fit(X,y)
plt.plot(df['Edad'],modelo.predict(X))
plt.scatter(df['Edad'],df['Salario'],marker="o",color="blue")
plt.title('Relación entre Edad y Salario')
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.grid(True);
```
